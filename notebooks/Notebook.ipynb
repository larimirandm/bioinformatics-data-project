```python
import pandas as pd
import matplotlib.pyplot as plt

# =========================
# Criar dataset fictício
# =========================

dados = {
    'diagnostico': [
        'Controle',
        'Infecção Viral',
        'Inflamação',
        'Infecção Bacteriana'
    ] * 5,

    'PCR_mg_L': [
        2,
        15,
        10,
        20
    ] * 5,

    'leucocitos_10e3_uL': [
        6,
        8,
        7,
        14
    ] * 5
}

# =========================
# Criar DataFrame
# =========================

df = pd.DataFrame(dados)

# =========================
# Visualizar primeiras linhas
# =========================

print(df.head())

# =========================
# Estatísticas descritivas
# =========================

print(df.describe())

# =========================
# Média de PCR por diagnóstico
# =========================

media_pcr = df.groupby('diagnostico')['PCR_mg_L'].mean()

print(media_pcr)

# =========================
# Gráfico
# =========================

media_pcr.plot(kind='bar')

plt.title('PCR médio por diagnóstico')
plt.ylabel('PCR mg/L')

plt.xlabel('Diagnóstico')

plt.show()
```

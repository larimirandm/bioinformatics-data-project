# Análise Exploratória de Dados Laboratoriais

```
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('../data/dados_bioinformatica_ficticios.csv')

df.head()
```

```
df.describe()
```

```
df.groupby('diagnostico')['PCR_mg_L'].mean().plot(kind='bar')
plt.title('PCR médio por diagnóstico')
plt.ylabel('PCR mg/L')
plt.show()
```
